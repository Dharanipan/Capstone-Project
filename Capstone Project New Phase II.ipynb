```python
import sqlite3                                       # Creating a new database and cursor.
dbcon=sqlite3.connect("Mycapstoneprojectdb.db")
cur=dbcon.cursor()
```

# Creating Table 1: Sno, MovieName, Director Name, Duration, genre, ratings

```python
cur.execute("create table A(Sno int, MovieName text primarykey, Director1 text,Director2 text, Director3 text, Duration int, Year int, Ratings real, Metascore int)")
```

```python
with open('Movielist.csv','r') as file:                # Inserting the values from csv file to table.
    result=0
    for row in file:
        cur.execute("insert into A values(?,?,?,?,?,?,?,?,?)",row.split(','))
        dbcon.commit()
        result+=1
print(result,"rows inserted")
```

```python
# 1)  Display all the details of movies created by directors Christopher and Matt Reeves.
results=cur.execute('select * from A where Director1 like "Christopher%" or "Matt Reeves"')
results.fetchall()
```

```python
# 2) Display all the details of movies with a duration of 140 minutes to 190 minutes.
results=cur.execute('select * from A where Duration between 140 and 190')
results.fetchall()
```

```python
# 3) Display all details of movies with ratings above 7 in ascending order.
results=cur.execute('Select * from A where Ratings>7 order by Ratings') 
results.fetchall()
```

```python
# 4) Display all movie names in descending order.
results=cur.execute("select MovieName from A order by MovieName desc")
results.fetchall()
```

```python
# 5) Display movie name starts with ‘P’ and their rating is greater than 7.
results=cur.execute('select MovieName from A where MovieName like "P%" and Ratings>7')
results.fetchall()
```

# Creating Table 2: Movie Name, stars, votes, Genre, Gross collection, popularity, Certification

```python
cur.execute("create table B(MovieName text, Stars1 text, Stars2 text, Stars3 text, Votes int, Genre1 text, Genre2 text, Genre3 text, Gross varchar, Popularity int, Certification varchar,foreign key(MovieName) references A(MovieName))")
```

```python
with open('Movie_full_details.csv','r') as file1:          # Insering values from csv file to table.
    result1=0
    for rows in file1:
        cur.execute("insert into B values(?,?,?,?,?,?,?,?,?,?,?)",rows.split(','))
        dbcon.commit()
        result1+=1
print(result1,"rows inserted")
```

```python
# 1) Display all movie names with star Arnold Schwarzenegger in ascending order.
results1=cur.execute('select MovieName,Stars1 from B where Stars1="Arnold Schwarzenegger" order by MovieName asc')
results1.fetchall()
```

```python
# 2) Display all details of the movie with the highest number of votes.
results1=cur.execute('select * from B where votes=(select max(Votes) from B)')
results1.fetchall()
```
