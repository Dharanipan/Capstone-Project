```python
import selenium
from selenium import webdriver                    # Importing necessary libraries.
import pandas as pd
import time
import warnings
warnings.filterwarnings("ignore")
from selenium.webdriver.common.by import By
from seleniumbase import Driver
from bs4 import BeautifulSoup
import requests
```

# Phase 1

```python
from seleniumbase import Driver

driver1=Driver(browser="chrome", headless=False)
driver1.get("https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1")
```

# Store the entire data in two different CSV files as per the given fields:

The first CSV file data contains : Sno, Movie Name, Director Name, Duration, year, ratings, Metascore

```python
SNo=[]
snotags=driver1.find_elements(By.XPATH,'//span[@class="lister-item-index unbold text-primary"]')
for i in snotags:
    SNo.append(i.text)

Movie_Name=[]
movienametags=driver1.find_elements(By.XPATH,'//h3[@class="lister-item-header"]/a')
for i in movienametags:
    Movie_Name.append(i.text)    

Duration=[]
durationtags=driver1.find_elements(By.XPATH,'//span[@class="runtime"]')
for i in durationtags:
    Duration.append(i.text.replace('min',''))
    
Year=[]
yeartags=driver1.find_elements(By.XPATH,'//span[@class="lister-item-year text-muted unbold"]')
for i in yeartags:
    Year.append(i.text.replace("("," ").replace(")"," "))
    
Ratings=[]
ratingstags=driver1.find_elements(By.XPATH,'//div[@class="inline-block ratings-imdb-rating"]/strong')
for i in ratingstags:
    Ratings.append(i.text)

links=[]
for i in driver1.find_elements(By.XPATH,'//h3[@class="lister-item-header"]/a'):
    links.append(i.get_attribute('href'))
Metascore=[]
Director1=[]
Director2=[]
Director3=[]
for i in links:
    driver1.get(i)
    time.sleep(10)
    try:
        metatags=driver1.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[2]/ul/li[3]/a/span/span[1]/span')
        Metascore.append(metatags.text)
    except:
        Metascore.append("NA")
    try:
        director1tags=driver1.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[2]/div/ul/li[1]/div/ul/li/a')
        Director1.append(director1tags.text)
    except:
        Director1.append("NA")
    try:
        director2tags=driver1.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[2]/div/ul/li[1]/div/ul/li[2]/a')
        Director2.append(director2tags.text)
    except:
        Director2.append("NA")
    try:
        director3tags=driver1.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[2]/div/ul/li[1]/div/ul/li[3]/a')
        Director3.append(director3tags.text)
    except:
        Director3.append("NA")  
```

```python
df1=pd.DataFrame({'S.No':SNo,'Movie Name':Movie_Name,'Director1':Director1,'Director2':Director2,'Director3':Director3,'Duration':Duration,'Year':Year,'Ratings':Ratings,'MetaScore':Metascore})
df1
```

```python
df1.to_csv("Movielist.csv",header=False,index=False)   # Saving the dataframe into csv file.
```

```python
driver1.close()
```

The second CSV file contains the following: Movie Name, stars, votes, Genre, Gross collection, popularity, Certification

```python
driver2=Driver(browser='chrome',headless=False)
driver2.get("https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1")
```

```python
links2=[]
for i in driver2.find_elements(By.XPATH,'//h3[@class="lister-item-header"]/a'):
    links2.append(i.get_attribute('href'))
```

```python
Votes=[]
Gross=[]
recruiter=[i.text.split('|') for i in driver2.find_elements(By.XPATH,'//p[@class="sort-num_votes-visible"]')]
for i in recruiter:
    try:
        Votes.append(i[0].replace("Votes: ","").replace(",",""))
    except:
        Votes.append('NA')
    try:
        Gross.append(i[1].replace("Gross: $",""))
    except:
        Gross.append('NA')

Movie_Name2=[]
Stars1=[]
Stars2=[]
Stars3=[]
Genre1=[]
Genre2=[]
Genre3=[]
Popularity=[]
Certification=[]

for i in links2:
    driver2.get(i)
    time.sleep(15)
    try:
        movietags=driver2.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/h1/span')
        Movie_Name2.append(movietags.text)
    except:
        Movie_Name2.append("NA")
    try:
        Star1tags=driver2.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[2]/div/ul/li[3]/div/ul/li[1]/a')
        Stars1.append(Star1tags.text)
    except:
        Stars1.append("NA")
    try:
        Star2tags=driver2.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[2]/div/ul/li[3]/div/ul/li[2]/a')
        Stars2.append(Star2tags.text)
    except:
        Stars2.append("NA")
    try:
        Star3tags=driver2.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[2]/div/ul/li[3]/div/ul/li[3]/a')
        Stars3.append(Star3tags.text)
    except:
        Stars3.append("NA")
    try:
        genre1tags=driver2.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[1]/div[2]/a[1]')
        Genre1.append(genre1tags.text)
    except:
        Genre1.append("NA")
    try:
        genre2tags=driver2.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[1]/div[2]/a[2]')
        Genre2.append(genre2tags.text)
    except:
        Genre2.append("NA")
    try:
        genre3tags=driver2.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[1]/div[2]/a[3]')
        Genre3.append(genre3tags.text)
    except:
        Genre3.append("NA")
    try:
        popularitytags=driver2.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[2]/div[2]/div/div[3]/a/span/div/div[2]/div[1]')
        Popularity.append(popularitytags.text.replace(",",""))
    except:
        Popularity.append("NA")
    try:
        certificationtags=driver2.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/ul/li[2]/a')
        Certification.append(certificationtags.text)
    except:
        Certification.append("NA")  
```

```python
print(len(Movie_Name2))
print(len(Stars1))
print(len(Stars2))
print(len(Stars3))
print(len(Votes))
print(len(Genre1))
print(len(Genre2))
print(len(Genre3))
print(len(Gross))
print(len(Popularity))
print(len(Certification))
```

```python
df2=pd.DataFrame({'Movie Name':Movie_Name2,'Stars1':Stars1,'Stars2':Stars2,'Stars3':Stars3, 'Votes':Votes, 'Genre1':Genre1, 'Genre2':Genre2,'Genre3':Genre3, 'Gross':Gross, 'Popularity':Popularity, 'Certification':Certification})
df2
```
